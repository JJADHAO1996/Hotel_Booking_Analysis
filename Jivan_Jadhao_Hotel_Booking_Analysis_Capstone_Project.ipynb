# **What is the best time of year to book a hotel room ?**

```python
#creating variable suitable_booking_month
suitable_booking_month=df["arrival_date_month"].value_counts()
```

```python
#checking the type of the variable
type(suitable_booking_month)
```

```python
#sorting the values of the varible
suitable_booking_month.sort_index(inplace=True)
```

```python
#print the variable
suitable_booking_month
```

```python
#using groupby function to make a group of two data columns from daaframe and store it in new variable
suitable_booking_year1 = df.groupby(['arrival_date_year','arrival_date_month'])['arrival_date_year'].count().unstack()
```

```python
#print the new variable
suitable_booking_year1
```

```python
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
#plot the graph using matplotlib
plt.figure(figsize = (15,5))
suitable_booking_year1.plot(kind='bar',stacked=False)
plt.title('Monthly Bookings in the year\n2015,2016,2017')
plt.ylabel('Total number of Bookings')
plt.xlabel('Month')
plt.legend()
plt.show()
```

```python
# Create the plot object
plt.figure(figsize = (15,5)) # controls size of plot
suitable_booking_month.plot(kind='bar', stacked=True)


plt.title('Combined Monthly Booking Traffic of\n2015,2016,2017')
plt.ylabel('Total number of Bookings')
plt.xlabel('Month')
plt.legend() #  shows labels in plot
plt.show()
```

**Observation:** In 'Monthly Bookings in the year 2015,2016,2017' bar graphical representation we can easily observed the hike in booking in the month of April, May and October which generally are the vacation period in India.But, In 'Combined Monthly Booking Traffic of n2015,2016,2017', It clearly replicate August month has Highest ever booking traffic. so we can say that Month of August, April, May and October will be suitable month for heavy booking for Hotel Owner. Rest of months should be great for customers to book cause of deman-suppy rule; Less demand Less Price.

# **How many Customers Checked-out and Canceled?**

```python
nctype = [df.reservation_status[df.reservation_status=='Check-Out'].count(),df.reservation_status[df.reservation_status=='Canceled'].count(),df.reservation_status[df.reservation_status=='No-Show'].count()]
ctype = ['Check-Out','Canceled','No-Show']
plt.pie(nctype,labels=ctype,startangle=90,colors=['orange','yellow','green'], autopct ='%1.1f%%')
plt.show()
```

**Observation:** We found that 63% customer have Checked-out, while 36% customers have Canceled the bookings. And we don't know the preference of 1.0% customres.
